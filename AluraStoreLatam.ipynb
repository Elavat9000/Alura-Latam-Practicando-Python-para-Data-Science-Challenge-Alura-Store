### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:,.2f} ")
```

```python
import matplotlib.pyplot as plt

ingresosTotales = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}

plt.figure(figsize=(10, 6))
plt.pie(
    ingresosTotales.values(),
    labels=ingresosTotales.keys(),
    autopct='%1.1f%%',
    startangle=75,
    colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'],
    explode=(0.05, 0.05, 0.05, 0.05),
    shadow=True
)

plt.title('Ingresos Totales por Tienda', pad=50, fontweight='bold')

leyenda = [f'{k}: ${v:,.0f} ' for k, v in ingresosTotales.items()]
plt.legend(leyenda, title="Ingresos", loc="best", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
todas_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, tienda in enumerate(todas_tiendas, start=1):
    ventas = tienda.groupby('Categoría del Producto').size().reset_index(name='Cantidad')
    print(f"\nTienda {i} - Ventas por categoría (ordenadas):")
    print(ventas.sort_values('Cantidad', ascending=False).to_string(index=False))
```

# 3. Calificación promedio de la tienda


```python
promedio_calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [
        tienda['Calificación'].mean(),
        tienda2['Calificación'].mean(),
        tienda3['Calificación'].mean(),
        tienda4['Calificación'].mean()
    ],
    'Número de Evaluaciones': [
        len(tienda),
        len(tienda2),
        len(tienda3),
        len(tienda4)
    ]
})

promedio_calificaciones = promedio_calificaciones.sort_values('Calificación Promedio', ascending=False)

print("Calificaciones promedio por tienda:")
print(promedio_calificaciones.to_string(index=False))
```

# 4. Productos más y menos vendidos

```python
def analizar_ventas_productos(tienda, nombre_tienda):
    ventas_por_producto = tienda['Producto'].value_counts().reset_index()
    ventas_por_producto.columns = ['Producto', 'Cantidad_Vendida']

    top_5 = ventas_por_producto.head(5)

    bottom_5 = ventas_por_producto.tail(5)

    print(f"\nAnálisis de ventas {nombre_tienda}:")
    print("\nTop 5 productos más vendidos:")
    print(top_5.to_string(index=False))

    print("\nTop 5 productos menos vendidos:")
    print(bottom_5.to_string(index=False))
    print("\n")
    print("-------------------------------------------------------")

    return top_5, bottom_5

top1, bottom1 = analizar_ventas_productos(tienda, "Tienda 1")
top2, bottom2 = analizar_ventas_productos(tienda2, "Tienda 2")
top3, bottom3 = analizar_ventas_productos(tienda3, "Tienda 3")
top4, bottom4 = analizar_ventas_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo Promedio Envío ': [
        tienda['Costo de envío'].mean(),  # Promedio para Tienda 1
        tienda2['Costo de envío'].mean(),  # Promedio para Tienda 2
        tienda3['Costo de envío'].mean(),  # Promedio para Tienda 3
        tienda4['Costo de envío'].mean()   # Promedio para Tienda 4
    ],
    'Número de Envíos': [
        len(tienda),  # Total de envíos en Tienda 1
        len(tienda2),  # Total de envíos en Tienda 2
        len(tienda3),  # Total de envíos en Tienda 3
        len(tienda4)   # Total de envíos en Tienda 4
    ]
})
```

```python
pd.options.display.float_format = '{:,.2f}'.format
print("Costo promedio de envío por tienda:")
print(costo_envio_promedio.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

ganancias = [
    (tienda['Precio'] - tienda['Costo de envío']).sum(),
    (tienda2['Precio'] - tienda2['Costo de envío']).sum(),
    (tienda3['Precio'] - tienda3['Costo de envío']).sum(),
    (tienda4['Precio'] - tienda4['Costo de envío']).sum()
]

df_ganancias = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ganancias': ganancias
})

df_ganancias = df_ganancias.sort_values('Ganancias', ascending=False)


plt.figure(figsize=(10, 6))
bars = plt.bar(df_ganancias['Tienda'],
               df_ganancias['Ganancias'],
               color=['#4CAF50', '#8BC34A', '#CDDC39', '#FFC107'])


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height:,.0f}',
             ha='center', va='bottom')


plt.title('Ganancias Totales por Tienda', pad=20, fontweight='bold')
plt.ylabel('Ganancias', labelpad=10)
plt.xlabel('Tienda', labelpad=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```
