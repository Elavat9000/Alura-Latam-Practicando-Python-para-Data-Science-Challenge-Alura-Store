#ðŸ”§ TransformaciÃ³n

```python
df_customer = pd.json_normalize(datos['customer'])
df_phone = pd.json_normalize(datos['phone'])
df_internet = pd.json_normalize(datos['internet'])
df_account = pd.json_normalize(datos['account'])

df_F = pd.concat([datos[['customerID','Churn']],df_customer, df_phone, df_internet, df_account], axis=1)
df_F.head()
```

```python
import numpy as np
```

```python
df_F['Charges.Total'] = df_F['Charges.Total'].replace(' ',np.nan)
df_F['Charges.Total'] = df_F['Charges.Total'].astype(float)
df_F.info()
```

```python
dataframe = pd.concat ([df_F['OnlineBackup'],df_F['OnlineSecurity'],df_F['DeviceProtection']], axis =1)
dataframe
```

```python
df_F.head()
```

```python
df_F[['Churn','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']] = df_F[['Churn','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']].applymap(lambda x: 1 if x == 'Yes' else 0)
```

```python
df_F[['Churn','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']] = df_F[['Churn','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']].astype(np.bool)
```

```python
df_F.info()
```

```python
df_F[['customerID','gender', 'InternetService', 'Contract', 'PaymentMethod']] = df_F[['customerID','gender', 'InternetService', 'Contract', 'PaymentMethod']].apply(lambda col: col.str.lower())
df_F.head()
```

```python
df_F[['customerID','gender', 'InternetService', 'PaymentMethod']] = df_F[['customerID','gender', 'InternetService', 'PaymentMethod']].applymap(lambda x: x.replace('-','').replace(',',''.strip()))
df_F.head()

df_F['Contract'] = df_F['Contract'].apply(lambda x: x.replace('-',' '))
df_F.head()
```
